```python
import pandas as pd
import numpy as np
```

```python
df_cust=pd.read_csv('AWCustomers.csv')
df_sales=pd.read_csv('AWSales.csv')
```

```python
print(df_cust.columns)
df_cust.head()
```

```python
df_merged=pd.merge(df_cust,df_sales,left_on='CustomerID',right_on='CustomerID',how='inner')
df_merged=df_merged.drop_duplicates(subset='CustomerID', keep='first')
df_merged.head()
```

```python
df_merged.info()
```

```python
columns_to_drop = [
    'CustomerID', 'Title', 'FirstName', 'MiddleName', 'LastName', 'Suffix','City','StateProvinceName',
    'AddressLine1', 'AddressLine2', 'PhoneNumber', 'BirthDate','PostalCode','TotalChildren','LastUpdated'
]
df_dropped = df_merged.drop(columns=columns_to_drop)
```

```python
df_dropped.head()
```

```python
Nominal = [
    'CountryRegionName',
    'Occupation',
    'Gender',
    'MaritalStatus'
]

Ordinal = [
    'Education'  
]

Ratio = [
    'HomeOwnerFlag',         
    'NumberCarsOwned',
    'NumberChildrenAtHome',
    'YearlyIncome',
    'AvgMonthSpend'
]

Interval = [
]

print("Nominal: ", Nominal)
print("Ordinal: ", Ordinal)
print("Ratio: ", Ratio)
print("Interval: ", Interval)
```

```python
df_dropped.isnull().sum()
```

```python
df_dropped['Gender']=df_dropped['Gender'].map({
    'M':1,
    'F':0
})
df_dropped['MaritalStatus']=df_dropped['MaritalStatus'].map({
    'M':1,
    'S':0
})

education_order = {
    'Partial High School': 0,
    'High School': 1,
    'Partial College': 2,
    'Bachelors': 3,
    'Graduate Degree': 4
}
df_dropped['Education'] = df_dropped['Education'].map(education_order)

df_dropped = pd.get_dummies(df_dropped, columns=['Occupation', 'CountryRegionName'])

from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
df_dropped[['YearlyIncome', 'AvgMonthSpend']] = scaler.fit_transform(df_dropped[['YearlyIncome', 'AvgMonthSpend']])
```

```python
df_dropped.describe()
```

```python
df_dropped.head()
```

```python
x = df_dropped.iloc[0]
y = df_dropped.iloc[1]

from sklearn.metrics import jaccard_score
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

x = x.to_numpy()
y = y.to_numpy()

x_bin = (x > np.median(x)).astype(int)
y_bin = (y > np.median(y)).astype(int)

def simple_matching(x, y):
    matches = np.sum(x == y)
    return matches / len(x)

jaccard = jaccard_score(x_bin, y_bin)

cos_sim = cosine_similarity([x], [y])[0][0]

smc = simple_matching(x_bin, y_bin)

print("Simple Matching Coefficient:", round(smc, 4))
print("Jaccard Similarity:", round(jaccard, 4))
print("Cosine Similarity:", round(cos_sim, 4))
```
